```python
# Importing Require Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, LassoCV, RidgeCV, ElasticNetCV, Ridge, Lasso
from sklearn.preprocessing import StandardScaler
import statsmodels.formula.api as sm
import scipy.stats as stats
```

```python
#Importing Dataset from sklearn
import sklearn
from sklearn.datasets import load_boston
boston = load_boston()
```

```python
df=pd.DataFrame(boston.data)
```

```python
df.head()
```

```python
#Checking Datatype of each Column & Here No any null values present.
df.info()
```

```python
df.describe()
```

```python
# Defining Features & Target Variable
x = df.iloc[:,:12]
y = df.iloc[:,12]
```

```python
x.head()
```

```python
# Converting x into Scaled Values.
scaler= StandardScaler()
x_scaled= scaler.fit_transform(x)
```

```python
# Checking Colinearity of features with the help of VIF 
from statsmodels.stats.outliers_influence import variance_inflation_factor
#variables= x_scaled
vif = pd.DataFrame()
vif["Features"] = x.columns
vif["VIF"] = [variance_inflation_factor(x_scaled, i) for i in range(x.shape[1])]
```

```python
vif
```

```python
x_scaled_df=pd.DataFrame(x_scaled)
x_scaled_df.head()
```

```python
#VIF Value is high for column name 7, 8. so droping these columns
x_scaled_df = x_scaled_df.drop([7,8], axis=1)
```

```python
x_scaled_df
```

```python
# spliting data into train & test
x_train,x_test,y_train,y_test= train_test_split(x_scaled_df,y,test_size=0.25)
```

```python
# Applying Linear Regression Model
regression = LinearRegression()
model= regression.fit(x_train, y_train)
```

```python
#Save model
import pickle
filename = 'Bostan_House_model.pickle'
pickle.dump(model, open(filename, 'wb'))
```

```python
y_pred = regression.predict(x_test)
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
def adj_r2(x,y):
    r2=regression.score(x,y)
    n=x.shape[0]
    p=x.shape[1]
    adjusted_r2 = 1-(1-r2)*(n-1)/(n-p-1)
    return(adjusted_r2)
```

```python
adj_r2(x_train,y_train)
```

```python
adj_r2(x_test,y_test)
```
